## cv2 automatically transforms the image file to numpy array
- imread로 사용하면 해당 path에 파일이 없어도 에러가 나오지 않기 때문에 무조건 type을 확인해주어야 한다.

```python
img = cv2.imread('../DATA/00-puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

## Matplotlib과 OpenCV의 색상 순서가 다르기 때문에 cvtColor()로 변환해주어야 한다

```python
# MATPLOTLIB --> RGB RED GREEN BLUE
# OPENCV ---> BLUE GREEN RED
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
fix_img.shape
```

## numpy와 cv2.resize의 xy 위치가 바뀐다

```python
new_img = cv2.resize(fix_img,(1000,400))
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.8
h_ratio = 0.2 # 20% of orignial height
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
new_img = cv2.flip(fix_img,0) # 0: 가로축, 1: 세로축, -1: 둘다
plt.imshow(new_img)
```

```python
type(fix_img)
```

```python
cv2.imwrite('totally_new.jpg',fix_img)
```
