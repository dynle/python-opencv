```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# copyright free iamge from pixabay
img = cv2.imread('walking.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_and_blur_face(img):
    img_copy = img.copy()
    face_rects = face_cascade.detectMultiScale(img_copy,scaleFactor=1.3,minNeighbors=5)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(img_copy,(x,y),(x+w,y+h),(0,0,255),5)
        
        roi = img_copy[y:y+h,x:x+w]
        blurred_roi = cv2.medianBlur(roi,39)
        
        img_copy[y:y+h,x:x+w] = blurred_roi
    
    return img_copy
```

```python
result = detect_and_blur_face(img)
```

```python
plt.imshow(result)
```
